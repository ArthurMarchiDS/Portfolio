Como falei, gerei um gráfico de dispersão em um mapa usando o Dataset do AirBnB,então vamos lá.

Acesse: Inside AirBnB

Pode pegar o Dataset de qualquer cidade disponível, o objetivo aqui é gerar o mapa apenas.

Irei utilizar o Dataset de Beijing, opte por escolher o arquivo listings.cvs ( Summary information and metrics for listings in Beijing (good for visualisations) por se tratar de uma versão mais resumida.

Vamos importar as bibliotecas necessárias:

```
import pandas as pd
import plotly.express as px
```

Vamos importar nosso Dataset, através do comando:

```
df = pd.read_csv('http://data.insideairbnb.com/china/beijing/beijing/2019-11-24/visualisations/listings.csv')
```

Note que usamos uma variável e logo em seguida usamos o Pandas e usamos a função 'read_csv' para carregar nosso Dataset.

Vamos usar a função 'head', para visualizar como está o Dataset.

```
df.head()
```

Note que temos várias colunas, incluindo as colunas de 'latitude' e 'longitude', e é justamente elas que usaremos para gerar nosso mapa.

Sem mais delongas, vamos lá.

Note que criamos uma variável chamada 'mapa' chamamos a biblioteca através da abreviação 'px' que foi utilizada ao importar a biblioteca e utilizamos o método 'scatter_mapbox' e usamos os parâmetros solicitados, note que os parâmetros 'lat' e 'lon' usamos as colunas 'latitude' e 'longitude' do nosso Dataset.

E na linha abaixo delimitamos o layout do nosso mapa.
Esse foi o resultado.

```
mapa = px.scatter_mapbox(df, lat='latitude', lon='longitude', color_discrete_sequence=['red'], width=1000, height=900)
mapa.update_layout(mapbox_style="open-street-map")
```

Podemos usar os mesmos parâmetros e apenas mudar o layout.

```
mapa = px.scatter_mapbox(df, lat='latitude', lon='longitude', color_discrete_sequence=['red'], width=1000, height=900)
mapa.update_layout(mapbox_style="stamen-terrain")
```
