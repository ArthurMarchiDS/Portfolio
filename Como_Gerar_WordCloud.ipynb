```
#Importar todas as bibliotecas necess√°rias
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS,ImageColorGenerator
```

```
#Importando o Dataset
df = pd.read_csv('http://dl.dropboxusercontent.com/s/fdzigk974zbx2de/listings.csv?dl=0')
```

```
df.head()
```

```
#Vamos usar s√≥ a coluna Summary, e iremos apagar todos os campos vazios
summary = df.dropna(subset = ['summary'], axis=0) ['summary']
```

```
#Juntando todas as palavras em uma √∫nica 'frase'
all_summary = ''.join(s for s in summary)
```

```
stopwords = set(STOPWORDS)
```

```
wordcloud = WordCloud(stopwords=stopwords).generate(all_summary)
```

```
#Gerando a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')

plt.show()
```

```
#Ainda cont√©m algumas stopwords, ent√£o vamos atualizar a lista.
stopwords = set(STOPWORDS)
stopwords.update(['da', 'meu', 'em', 'voc√™', 'de', 'ao', 'os'])
```

```
#Alterando a cor de fundo.
wordcloud = WordCloud(stopwords=stopwords, background_color='white').generate(all_summary)
```

```
#Gerando a imagem final
fig, ax = plt.subplots(figsize=(10,6))

ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
plt.show()
```
